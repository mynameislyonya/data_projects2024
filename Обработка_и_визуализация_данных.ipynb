**1)	Чтение и запись файлов с помощью pandas.**



a.	Прочитать 2 файла «1. data1» и «1. data2» (в каждом файле 2 листа).

b.	Объединить данные из файлов в один датафрейм со столбцами:
X1, x2, y

c.	Записать исходный датафрейм в csv файл с разделителями «,».

```python
import pandas as pd

df1_1 = pd.read_excel('1. data1.xlsx', sheet_name='фичи1')
df2_1 = pd.read_excel('1. data2.xlsx', sheet_name='ответы1')

df1_2 = pd.read_excel('1. data1.xlsx', sheet_name='фичи2')
df2_2 = pd.read_excel('1. data2.xlsx', sheet_name='ответы2')

df1 = pd.concat([df1_1, df2_1], axis=1, ignore_index=True)
df2 = pd.concat([df1_2, df2_2], axis=1, ignore_index=True)

df1.columns = ['X1', 'x2', 'y']
df2.columns = ['X1', 'x2', 'y']

df = pd.concat([df1, df2], axis=0, ignore_index=True)

df.to_csv('merged_data.csv', index=False, sep=',')

print("Данные успешно объединены и записаны в файл merged_data.csv.")
```

**2)	Работа с типами данных.**

a.	Прочитать первые 4 столбца файла «2. Data», пропустить при этом первые 3 строчки

b.	Столбец «№» – преобразовать в тип float

c.	Столбцы «дата1» и «дата2» – преобразовать в формат DateTime

d.	Создать столбец «дней до сегодня» - считающий кол-во дней от «дата1» до сегодняшней даты.


```python
import pandas as pd
from datetime import datetime

df1 = pd.read_excel('2. Data.xlsx', skiprows=3, usecols=[0, 1, 2, 3])

df1['№'] = df1['№'].astype(float)

df1['дата1'] = pd.to_datetime(df1['дата1'], format='%d.%m.%Y')
df1['дата2'] = pd.to_datetime(df1['дата2'], format='%d.%m.%Y')

today = datetime.now().date()
df1['дней до сегодня'] = (today - df1['дата1'].dt.date).dt.days
```

**3)	 Предобработка текста и визуализация данных.**

a.	Прочитать файл, в котором хранятся данные о продажах различных моделей автомобилей с указанием цвета машины. Назвать колонки «Марка и модель», «Цвет» и «Объем продаж, шт»

b.	Предобработать данные (привести к единому виду)

c.	Преобразовать форматы данных, если требуется

d.	Задача: визуализируйте одним интерактивным графиком ответ на вопрос «Какой цвет наиболее востребован у каждой из моделей проданных авто?»

e.	Сделайте выводы по графику.


```python
import pandas as pd

df3 = pd.read_excel('3. Data.xlsx', names=['Марка и модель', 'Цвет', 'Объем продаж, шт'])
df3['Объем продаж, шт'] = df3['Объем продаж, шт'].replace('-', 0)
df3['Объем продаж, шт'] = df3['Объем продаж, шт'].astype(int)
df3['Цвет'] = df3['Цвет'].str.lower().str.title()
df3['Цвет'] = df3['Цвет'].fillna('Неизвестно')
df3['Цвет'] = df3['Цвет'].replace('Пусто', 'Неизвестно')
df3['Цвет'] = df3['Цвет'].str.strip()
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.offline import plot

color_mapping = {
    'Черный': ('Black', 'rgb(0, 0, 0)'),
    'Серый': ('Gray', 'rgb(128, 128, 128)'),
    'Красный': ('Red', 'rgb(255, 0, 0)'),
    'Белый': ('White', 'rgb(255, 255, 255)'),
    'Неизвестно': ('Unknown', 'rgb(192, 192, 192)'),
    'Синий': ('Blue', 'rgb(0, 0, 255)'),

}

model_color_sales = df3.groupby(['Марка и модель', 'Цвет'])['Объем продаж, шт'].sum().reset_index()

popular_colors = model_color_sales.groupby('Марка и модель')['Объем продаж, шт'].idxmax()
popular_model_color_sales = model_color_sales.loc[popular_colors]

popular_model_color_sales['Цвет'] = popular_model_color_sales['Цвет'].map(color_mapping)

fig = go.Figure(data=[go.Bar(
    x=popular_model_color_sales['Марка и модель'],
    y=popular_model_color_sales['Объем продаж, шт'],
    marker=dict(color=[color[1] for color in popular_model_color_sales['Цвет']]),
    hovertext=[f"Цвет: {color[0]}, RGB: {color[1]}" for color in popular_model_color_sales['Цвет']],  # Add hover text for color labels
)])

fig.update_layout(
    title='Объем продаж по популярным моделям и цветам',
    xaxis_title='Марка и модель',
    yaxis_title='Объем продаж',
)

fig.show()
```
